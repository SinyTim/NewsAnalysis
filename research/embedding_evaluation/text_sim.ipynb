```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist

from pathlib import Path
```

```python
path_corpus = Path('../../data/corpora/tutby_126784.csv')

# path_embedding = Path('../../data/emb/emb_clean_tutby_126784_header_d2v.npy')
path_embedding = Path('../../data/emb/tutby_126784_doc_rubert.npy')
```

```python
data = pd.read_csv(path_corpus)

print(data.shape)
display(data.head(5))
```

```python
with open(path_embedding, 'rb') as file:
    embeddings = np.load(file)
    
print(embeddings.shape)
```

```python
for k in range(10):
    dists = cdist(embeddings[k:k+1], embeddings, metric='cosine')[0]
    index = np.nanargmin(dists[k+1:])
    print(k)
    print(data['header'][k])
    print('-' * 20)
    print(data['header'][index + k+1])
    print('-' * 20)
```
